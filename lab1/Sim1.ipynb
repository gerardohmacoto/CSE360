# Building a Robot Simulator

```python
#perfect sensor
def sense(x):
    return x
```

## Euler Integration

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

## Control Policy

```python
def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = sin(t)
    uy = cos(t)
    
    return array([ux, uy])
```

## Running Simulator

```python
tf = 3. # Final time
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([2., 1.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```

```python
grid()
plot(x_log[:,0], x_log[:,1])
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation
# from JSAnimation import IPython_display    
from IPython.display import HTML
    


fig, ax = plt.subplots()

def animate(t):
    ax.clear()
    
    # Path
    plot(x_log[:,0], x_log[:,1], 'r--')
    
    # Initial conditions
    plot(x_log[t,0], x_log[t,1], 'bo')
    
    

anim = animation.FuncAnimation(fig, animate, frames=len(time), interval=60)

HTML(anim.to_jshtml())
```

## Control Policy for Ellipse

```python
def control_ellipse(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = 4*cos(t)
    uy = 2*sin(t)
    
    return array([ux, uy])
```

## Running Simulation for Ellipse

```python
tf_e = 7.
Δt_e = 0.1
time_e = linspace(0,tf_e, int(tf_e / Δt_e)+1)

x_e = array([3.,0.])
x_e_log = [copy(x_e)]

for t in time_e:
    y_e = sense(x_e)
    u_e = control_ellipse(t,y_e)
    x_e = simulate(Δt_e, x_e, u_e)
    x_e_log.append(copy(x_e))
x_e_log = array(x_e_log)
# print("x_e_log: ")
# print(x_e_log)
```

```python
print("x_e_log grid: ")
grid()
plot(x_e_log[:,0], x_e_log[:,1])
```
