```python
#sensor
def sense_x(x):
    return x
def sense_y(y):
    return y
```

```python
def simulate(ct,x,y,u):
    x = x + (ct * u[0])
    y = y + (ct * u[1])
    return array([x,y])

# def simulate_y(ct,y,u):
#     y = y + (ct * u)
#     return y
```

```python
def control(x,y):
    #p-controller constant
    k = 20
    # Desired position (x,y) => (2,10) (initial pos ==> (4,5))     
    xd = 2
    yd = 10     
    # multiply -1 to get positive values
    ux = k*(x-xd) * -1
    uy = k*(y-yd) * -1
    return array([ux,uy])
```

```python
ft = 7. #final time
ct = 0.1 #Time step 
time = linspace(0.,ft, int(ft/ct)+1) #time interval

#initial Conditions:
##### equivalent below for x and y #######
            # x = array([4.,5.])
            # x_log = [copy(x)]


# initial position (4,5)
x = 4
x_log = [copy(x)]

y = 5
y_log = [copy(y)]


for t in time:
    x = sense_x(x)
    y = sense_y(y)
    u = control(x,y)
    stored = simulate(ct,x,y,u)
    x = stored[0]
    y = stored[1]
    x_log.append(copy(x))
    y_log.append(copy(y))

x_log = array(x_log)
y_log = array(y_log)

# print(x_log)
# print("\n\n")
# print(y_log)
```
