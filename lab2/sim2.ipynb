# Programming the Freenove Robot

```python
import time
from Led import *
from Motor import *
from Buzzer import *

led=Led()
PWM=Motor()
buzzer=Buzzer()

def forward_motor():
    try:
        PWM.setMotorModel(1000,1000,1000,1000)
        print("Forward")
        time.sleep(2)
        PWM.setMotorModel(0,0,0,0)
    except KeyboardInterrupt:
        PWM.setMotorModel(0,0,0,0)
        print("End of program!")
def left_motor():
    try:
        PWM.setMotorModel(-500,-500,1750,1450)
        print("Left")
        time.sleep(2)
        PWM.setMotorModel(0,0,0,0)
    except KeyboardInterrupt:
        PWM.setMotorModel(0,0,0,0)
        print("End of program!")

def LED_on_red():
    try:
        led.ledIndex(0x01, 255,0,0)
        time.sleep(1)
        led.colorWipe(led.strip, Color(0,0,0))
    except KeyboardInterrupt:
        led.colorWipe(led.strip, Color(0,0,0))
        print("\nEnd of program!")
def LED_on_green():
    try:
        led.ledIndex(0x02, 0,255,0)
        time.sleep(1)
        led.colorWipe(led.strip, Color(0,0,0))
    except KeyboardInterrupt:
        led.colorWipe(led.strip, Color(0,0,0))
        print("\nEnd of program!")
def LED_on_blue():
    try:
        led.ledIndex(0x03, 0,0,255)
        time.sleep(1)
        led.colorWipe(led.strip, Color(0,0,0))
    except KeyboardInterrupt:
        led.colorWipe(led.strip, Color(0,0,0))
        print("\nEnd of program!")
def LED_on_yellow():
    try:
        led.ledIndex(0x04,255,255,0)
        time.sleep(1)
        led.colorWipe(led.strip, Color(0,0,0))
    except KeyboardInterrupt:
        led.colorWipe(led.strip, Color(0,0,0))
        print("\nEnd of program!")
def buzzer_on():
    try:
        buzzer.run('1')
        time.sleep(1)
        print('1S')
        buzzer.run('0')
        print("Buzzer stopped!")
    except KeyboardInterrupt:
        buzzer.run('0')
        print("\nEnd of program!")
        
def sysRun():
    #forward 2 seconds
    forward_motor()
    left_motor()
    LED_on_red()
    forward_motor()
    left_motor()
    LED_on_blue()
    forward_motor()
    left_motor()
    LED_on_green()    
    forward_motor()
    left_motor()
    LED_on_yellow()
    buzzer_on()
    

if __name__ == '__main__':

    print ('Program is starting ... ')
    import sys
    if len(sys.argv)<2:
        print ("Parameter error: Please assign the device")
        exit() 
    if sys.argv[1] == 'sysRun':
        sysRun()
```
